```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli", device=0)
```

```python
sequence_to_classify = "one day I will see the world"
candidate_labels = ['positive', 'negative']
classifier(sequence_to_classify, candidate_labels)
#{'labels': ['travel', 'dancing', 'cooking'],
# 'scores': [0.9938651323318481, 0.0032737774308770895, 0.002861034357920289],
# 'sequence': 'one day I will see the world'}
# add result to dataframe
# normalize score to range from -1 to 1
# for every song, get classification in these ranges:
# 1st axis: positive to negative
# 2nd axis: unique to boring
# 3rd axis: formal to informal
```

```python
def classify_song(classifier, song, labels):
    candidate_labels = labels
    result = classifier(song, candidate_labels)
    # add and normalize scores
    score = result['scores'][0] - result['scores'][1]
    return score
```

```python
import pandas as pd
with open('dataFile/dataFile.txt', 'r') as file:
    lines = file.readlines()
    for index, line in enumerate(lines):
        lines[index] = line.strip()

dict_result = {'positivity': [], 'uniqueness': [], 'formality': []}

for i, line in enumerate(lines):
    result = classifier(line, candidate_labels)
    positivity = classify_song(classifier, line, ['positive', 'negative'])
    uniqueness = classify_song(classifier, line, ['unique', 'boring'])
    formality = classify_song(classifier, line, ['formal', 'informal'])
    dict_result['positivity'].append(positivity)
    dict_result['uniqueness'].append(uniqueness)
    dict_result['formality'].append(formality)
    # save dictionary to csv every 100 songs
    if i % 100 == 0 and i != 0:
        df_result = pd.DataFrame(dict_result)
        df_result.to_csv(f'dataFile/dataFile{i}.csv')
        dict_result = {'positivity': [], 'uniqueness': [], 'formality': []}
    print(f'{i} songs processed')
```
